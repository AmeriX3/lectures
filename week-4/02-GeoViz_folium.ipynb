````python
# Iterating over rows in a df

# enumerate for lists = iterrows for dfs
# they both get the index and the element

# FOR MARKERS, I NEED TWO THINGS (at least)

# 1. Location
# 2. Icon


for index, row in df.iterrows():
    
     # 1. Location (and some other things like the tooltip)
    district = {"location": [row["lat"], row["lon"]], "tooltip": row["distrito"]}

    
    # 2. Icon based on the value
    if row["festividad"] == "Laborable": # work day
        icon = Icon (
            color="red",
            prefix="fa",
            icon="laptop",
            icon_color = "black")
         
    elif row["festividad"] == "Vispera": #eve day
            icon = Icon (
                color="blue",
                prefix="fa",
                icon="glass",
                icon_color = "black")
        
    elif row["festividad"] == "Festivo":
            icon = Icon (
                color="darkpurple",
                prefix="fa",
                icon="bug",
                icon_color = "black")
            
    else:
            icon = Icon (
                color="darkpurple",
                prefix="fa",
                icon="paypal",
                icon_color = "yellow")
        
    new_marker = Marker(**district, icon=icon)
    new_marker.add_to(madrid_map)
madrid_map
````

````python
madrid_map.save("bikersmadrid")
````

##  Morning bikers will be safer  🚴

Let's check how many categories there are in the time column, to make a heat map. First we make a map with the morning schedule, and then we add the afternoon schedule with the Layer Control.
Here we go 🔥

### We create the basic map to which we then add the layers

````python
madrid_map_day = Map(location = [mad_lat, mad_long], zoom_start=15)
madrid_map_day
````

### Subset the dataframe per layer

````python
morning = df[df["horario"] == "Mañana"]
morning.head(5)
````

### We create the FeatureGroup

````python
morning_group = folium.FeatureGroup(name="Mañana")
````

### Create the HeatMap layer and add it to the FeatureGroup

````python
HeatMap(data=morning[["lat", "lon"]], radius=15,
       gradient = {0.4: 'yellow', 0.65: 'orange', 1: 'white'}).add_to(morning_group)
````

### Add the FeatureGrup with heatmap and name to the map

````python
morning_group.add_to(madrid_map_day)
````

````python
madrid_map_day
````

### Adding the evenings

````python
evening = df[df["horario"]=="Tarde"]
evening.sample()
````

````python
evening_group = folium.FeatureGroup("Tarde")
HeatMap(data=evening[["lat", "lon"]], radius = 15, gradient = {0.4: 'blue', 0.65: 'lime', 1: 'red'}).add_to(evening_group)
evening_group.add_to(madrid_map_day)
````

````python
madrid_map_day
````

````python
madrid_map_day
````

### Night layer

````python
# 0. Create a base map

# 1. Create sub-df for just nights
# 2. Create a FeatureGroup for nights
# 3. Create the heatmap

# 4. We add the FeatureGroup to the base map
# 5. Add the legend for Layers
````

````python
#1. Subdf

df_nights = df[df["horario"] == "Noche"]
df_nights.sample()
````

````python
# 2. FeatureGroup

night_group = folium.FeatureGroup(name="Noche")
````

````python
# 3. Create the heatmap

HeatMap(data=df_nights[["lat", "lon"]], radius=15).add_to(night_group)
````

````python
# 4. We add the FeatureGroup to the base map
night_group.add_to(madrid_map_day)
````

````python
# 5. Layers
````

````python
folium.LayerControl(collapsed=False).add_to(madrid_map_day)
````

````python
madrid_map_day
````
